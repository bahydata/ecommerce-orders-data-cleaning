## Detected Issues

- Missing values in: shipment_date,customer_email,delivery_city
- Duplicates: 4 rows duplicates
- Data type issues: order_date,shipment_date-> date , unit_price->num
- Columns contain inconsistent categorical values due to case variation (e.g., "Delivered", "DELIVERED", "delivered").
- Columns contain inconsistent date formats (e.g., "2025-11-02", "15-Nov-2025", "02/02/2025").

## Cleaning Decisions

- Drop duplicates
- Handle missing values
- Normalize text (strip + lower + title)
- Convert data types

```python
df=df.drop_duplicates()
```

```python
for col in df.select_dtypes(include="object"):
    df[col] = df[col].str.strip().str.lower().str.title()

df.head()
```

```python
df.columns = df.columns.str.strip().str.lower()
df.head()
```

```python
#Convert dates, nums
df["order_date"] = pd.to_datetime(df["order_date"],format="mixed",dayfirst=True, errors="coerce")
df["shipment_date"] = pd.to_datetime(df["shipment_date"],format="mixed",dayfirst=True, errors="coerce")
df.info()
```

```python
df["unit_price"]=df["unit_price"].replace(r"[^\d.]","",regex=True).pipe(pd.to_numeric)
df["unit_price"].isna().sum()
```

```python
#handle with Outliers
df=df[ df["shipment_date"]>=df["order_date"]]
df=df[ df["quantity"] > 0 ]
df=df[ df["unit_price"] > 0 ]
df = df[(df['order_status'] != "Delivered") | (df['shipment_date'].notna())]
df = df[(df['order_status'] != "Shipped") | (df['shipment_date'].notna())]
```

```python
df["customer_email"]=df["customer_email"].fillna("UNKOWN_EMAIL")
df["delivery_city"]=df["delivery_city"].fillna("UNKOWN_city")
```
